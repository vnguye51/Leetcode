## 2 Keys Keyboard

 Initially on a notepad only one character 'A' is present. You can perform two operations on this notepad for each step:

    Copy All: You can copy all the characters present on the notepad (partial copy is not allowed).
    Paste: You can paste the characters which are copied last time.

Given a number n. You have to get exactly n 'A' on the notepad by performing the minimum number of steps permitted. Output the minimum number of steps to get n 'A'.

Example 1:

Input: 3
Output: 3
Explanation:
Intitally, we have one character 'A'.
In step 1, we use Copy All operation.
In step 2, we use Paste operation to get 'AA'.
In step 3, we use Paste operation to get 'AAA'.

* This sounds similar to a bitshift problem. We can either double the number of A's or use the previously recorded copy to paste A's on.

* The fastest we can create A's is to copy

* This appears to be a possible recursion? 

* This is also based on multiples, once we have the largest multiple of a number we can just paste it several times.
    * multiples of 12 are 1,12 2,6, 3,4. 
        * We can paste 1, 12 times ( copy A , paste A * 11, 12 steps(
        * We can reach 2 and then paste it 6 times ( copy A, paste A, copy A, paste A * 5, 8 steps)
        * We can reach 3 and then paste it 4 times(copy A, paste A, paste A, copy AAA, paste AAA * 3 , 7 steps)
        
* Therefore we can build this based on the multiples of a number. It might be likely that we might have to check each multiple pair to find one that matches 


----

* Create a memo that lasts the number of steps to reach a certain number, store this in the class
* Create a memo that stores the multiples of each number

* If a number is prime then the number of steps to get that number is itself
    * Initialize the memo with all primes from 1-1000(the range of n)
* For a given n check if its already in the memo. if it is return that number. If it isn't loop through its multiples
    use the memo to check how many steps are necessary to generate that number, then the number of steps to get to n is the (# of steps of multiple) + (its multiple pair) + 1(copying it)


```python
import math as math
class Solution:
    def __init__(self):
        self.memo = {}
        self.multiples = {}
        for n in range(2,1001):
            if self.checkPrime(n):
                self.memo[n] = n
            else:
                self.memo[n] = -1
            self.multiples[n] = self.getMultiples(n)
        self.memo[1] = 0
    def minSteps(self, n):
        if self.memo[n] != -1:
            return self.memo[n]
        Min = 10e100
        for multiple in self.multiples[n]:
            candidate = self.minSteps(multiple) + n//multiple 
            if candidate < Min:
                Min = candidate
        self.memo[n] = Min
        return Min
        
    def checkPrime(self,n):
        if n == 2 or n == 1:
            return True
        for i in range(2,math.floor(n**0.5)+1):
            if n%i == 0:
                return False
        return True
    
    def getMultiples(self,n):
        multiples = []
        for i in range(2,math.floor(n**0.5)+1):
            if n%i == 0:
                multiples.append(i)
                multiples.append(n//i)
        return multiples
```

```python
sol = Solution()
for i in range(1,1001):
    print(sol.minSteps(i),i)
print(sol.minSteps(6))
```
