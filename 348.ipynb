# Design Tic-Tac-Toe

Design a Tic-tac-toe game that is played between two players on a n x n grid.

You may assume the following rules:

    A move is guaranteed to be valid and is placed on an empty block.
    Once a winning condition is reached, no more moves is allowed.
    A player who succeeds in placing n of their marks in a horizontal, vertical, or diagonal row wins the game.

* We need a reset function
* We need a way to check for a win
    * We can check for a win
* We need a move command
    * move needs to return "1" or "2" if a player wins, and "0" if a player doesn't
    * Moves are disabled after a win condition is reached
    * Once a move is made check if a win is possible by checking if all moves in that diagonal, row, and col are filled

```python
class TicTacToe:
    def __init__(self, n):
        self.turn = 1
        #1 for x's turn, -1 for o's turn
        self.board = [["_" for i in range(0,n)] for j in range(0,n)]
        self.win = False
        """
        Initialize your data structure here.
        :type n: int
        """
        

    def move(self, row, col, player):
        """
        Player {player} makes a move at ({row}, {col}).
        @param row The row of the board.
        @param col The column of the board.
        @param player The player, can be either 1 or 2.
        @return The current winning condition, can be either:
                0: No one wins.
                1: Player 1 wins.
                2: Player 2 wins.
        :type row: int
        :type col: int
        :type player: int
        :rtype: int
        """
        self.board[row][col] = player
        return self.checkWin(row,col,player)
    
    def checkWin(self,row,col,player):
        flag = True
        for i in range(0,len(self.board)):
            if self.board[row][i] != player:
                flag = False
                break
        if flag:
            return player
        flag = True
        for i in range(0,len(self.board)):
            if self.board[i][col] != player:
                flag = False
                break
        if flag:
            return player
        flag = True
        if row-col == 0:
            for i in range(0,len(self.board)):
                if self.board[i][i] != player:
                    flag = False
                    break
        if flag:
            return player
        flag = True
        if row+col == len(self.board):
            for i in range(0,len(self.board)):
                if self.board[i][len(self.board)-1-i] != player:
                    flag = False
                    break
        if flag:
            return player
        return 0


# Your TicTacToe object will be instantiated and called as such:
# obj = TicTacToe(n)
# param_1 = obj.move(row,col,player)
```
