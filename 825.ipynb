## Friends of Appropriate Ages

Some people will make friend requests. The list of their ages is given and ages[i] is the age of the ith person. 

Person A will NOT friend request person B (B != A) if any of the following conditions are true:

    age[B] <= 0.5 * age[A] + 7
    age[B] > age[A]
    age[B] > 100 && age[A] < 100

Otherwise, A will friend request B.

Note that if A requests B, B does not necessarily request A.  Also, people will not friend request themselves.

How many total friend requests are made?

* Brute force approach

* For each person perform each check on every other person
* a person will always friend request themselves so subtract len(ages) from the Sum
* If the array is sorted we only have to look up to that index

* If A friend requests B then B will not friend request A unless they are the same age and we can remove them from the list to check


* Once a person is found that passes the test : age[B] <= 0.5 * age[A] + 7
    * Mark its index then search for the first value for age[B] > age[A] 
        * The number of requests are those two values subtracted from each other 

```python
import time as time
def numFriendRequests(ages):
    ages.sort()
    Sum = 0
    marker = 0
    marked = False
    for i,n in enumerate(ages):
        marked = False
        for j in range(marker,len(ages)):
            if marked == False:
                if not ages[j] <= 0.5* n + 7:
                    marked = True
                    marker = j
            if marked == True:
                if ages[j] > n:
                    Sum = Sum + (j - marker) - 1
                    break
        if j == len(ages)-1:
            print(i,j,marker)
            Sum = Sum + (j - marker)
    return Sum
```
